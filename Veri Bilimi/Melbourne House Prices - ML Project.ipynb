###  Eksik Gözlem Analizi bitti

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale, StandardScaler
from sklearn import model_selection
from sklearn.linear_model  import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn import neighbors
from sklearn.svm import SVR
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df = df.select_dtypes(include=["float64","int64"])
df
```

```python
y= df["Price"]
X = df.drop(["Price"],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

#  KNN

```python
knn_params={"n_neighbors":range(1,31)}
knn = KNeighborsRegressor()
knn_cv_model = GridSearchCV(knn, knn_params, cv=10).fit(X_train,y_train)
knn_cv_model.best_params_
```

```python
knn_tuned = KNeighborsRegressor(n_neighbors=knn_cv_model.best_params_["n_neighbors"]).fit(X_train,y_train)
y_pred = knn_tuned.predict(X_test)
print(f"MAE=>{mean_absolute_error(y_test, y_pred)}\nMSE=>{mean_squared_error(y_test,y_pred)}\nRMSE=>{np.sqrt(mean_squared_error(y_test,y_pred))}\nR^2 Score=>{r2_score(y_test, y_pred)}")
```

#  SVR Destek Vektör Makineleri

```python
svr_model = SVR(kernel="linear")
svr_params={"C":[0.1,0.5,1.3]}
```

```python
svr_cv_model = GridSearchCV(svr_model,svr_params,cv=5, verbose=2, n_jobs=-1).fit(X_train,y_train)
svr_cv_model.best_params_
```

```python
svr_tuned = SVR(kernel="linear", C=svr_cv_model.best_params_["C"]).fit(X_train,y_train)
y_pred = svr_tuned.predict(X_test)
print(f"MAE=>{mean_absolute_error(y_test, y_pred)}\nMSE=>{mean_squared_error(y_test,y_pred)}\nRMSE=>{np.sqrt(mean_squared_error(y_test,y_pred))}\nR^2 Score=>{r2_score(y_test, y_pred)}")
```

# Yapay Sinir Ağları

```python
mlp_model = MLPRegressor().fit(X_train,y_train)
mlp_params = {"alpha":[0.1, 0.01, 0.02, 0.001, 0.0001],
             "hidden_layer_sizes":[(10,20),(5,5), (100,100)]}
mlp_cv_model = GridSearchCV(mlp_model, mlp_params, cv = 10).fit(X_train,y_train)
mlp_cv_model.best_params_
```

```python
#Final Modeli
mlp_tuned = MLPRegressor(alpha=0.001, hidden_layer_sizes =(100,100)).fit(X_train,y_train)
y_pred = mlp_tuned.predict(X_test)
print(f"MAE=>{mean_absolute_error(y_test, y_pred)}\nMSE=>{mean_squared_error(y_test,y_pred)}\nRMSE=>{np.sqrt(mean_squared_error(y_test,y_pred))}\nR^2 Score=>{r2_score(y_test, y_pred)}")
```

```python
cart_tuned = DecisionTreeRegressor(max_depth=10, min_samples_split=40).fit(X_train, y_train)
y_pred = cart_tuned.predict(X_test)
print(f"MAE=>{mean_absolute_error(y_test, y_pred)}\nMSE=>{mean_squared_error(y_test,y_pred)}\nRMSE=>{np.sqrt(mean_squared_error(y_test,y_pred))}\nR^2 Score=>{r2_score(y_test, y_pred)}")
```

# Random Forests

```python
rf_params = {
    "max_depth" : [5,8,10],
    "max_features": [2,5,10],
    "n_estimators": [200,500,1000,2000],
    "min_samples_split": [2,10,80,100]
}
```

```python
rf_model = RandomForestRegressor(random_state=42).fit(X_train,y_train)
rf_cv_model = GridSearchCV(rf_model,rf_params, cv=10, verbose=3, n_jobs=-1).fit(X_train,y_train)
rf_cv_model.best_params_
```

```python
rf_tuned = RandomForestRegressor(n_estimators=200 ,max_depth=8, max_features=2, min_samples_split=2).fit(X_train,y_train)
y_pred= rf_tuned.predict(X_test)
print(f"MAE=>{mean_absolute_error(y_test, y_pred)}\nMSE=>{mean_squared_error(y_test,y_pred)}\nRMSE=>{np.sqrt(mean_squared_error(y_test,y_pred))}\nR^2 Score=>{r2_score(y_test, y_pred)}")
```

# GBM

```python
gbm_params = {
    "learning_rate": [0.001,0.01,0.1],
    "subsample": [0.5,1.0,1.5],
    "n_estimators": [100,200,300],
    "loss": ["absolute_error", "huber","quantile","squared_error"],
    "max_depth":[3,5,8]    
}
gbm_model = GradientBoostingRegressor()
gbm_cv_model = GridSearchCV(gbm_model,gbm_params,cv=5, n_jobs=-1, verbose=2).fit(X_train,y_train)
gbm_cv_model.best_params_
```

```python
gbm_tuned = GradientBoostingRegressor(max_depth=3, n_estimators=200, loss='squared_error', subsample=0.5, learning_rate=0.1).fit(X_train,y_train)
y_pred = gbm_tuned.predict(X_test)
print(f"MAE=>{mean_absolute_error(y_test, y_pred)}\nMSE=>{mean_squared_error(y_test,y_pred)}\nRMSE=>{np.sqrt(mean_squared_error(y_test,y_pred))}\nR^2 Score=>{r2_score(y_test, y_pred)}")
```
